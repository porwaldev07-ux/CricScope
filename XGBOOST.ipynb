```python
matches=pd.read_csv("matches.csv")
deliveries=pd.read_csv('deliveries.csv')
```

```python
TEAM_ALIAS = {
    "Delhi Daredevils":            "Delhi Capitals",
    "Kings XI Punjab":             "Punjab Kings",
    "Royal Challengers Bengaluru": "Royal Challengers Bangalore",
    "Deccan Chargers":             "Sunrisers Hyderabad",
}

for col in ["team1", "team2", "winner"]:
        matches[col] = matches[col].replace(TEAM_ALIAS)
```

```python
matches['date']=pd.to_datetime(matches['date'],format='%d-%m-%Y')
```

```python
matches['year']=matches['date'].dt.year
```

```python
total_score_df = deliveries.groupby(['match_id', 'inning'])['total_runs'].sum().reset_index()
total_score_df = total_score_df[total_score_df['inning'] == 1]
total_score_df['target_score'] = total_score_df['total_runs'] + 1
total_score_df = total_score_df[['match_id', 'target_score']]
```

```python
match_df = matches.merge(total_score_df, left_on='id', right_on='match_id')
match_df = match_df[['match_id', 'team1', 'team2', 'winner', 'win_by_runs', 'win_by_wickets', 'target_score']]
```

```python
delivery_df = deliveries[deliveries['inning'] == 2].copy()
```

```python
delivery_df = delivery_df.merge(match_df, on='match_id')
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs'].cumsum()
delivery_df['runs_left'] = delivery_df['target_score'] - delivery_df['current_score']
```

```python
delivery_df['runs_left'] = delivery_df['runs_left'].clip(lower=0)
```

```python
delivery_df['balls_bowled'] = (delivery_df['over'] - 1) * 6 + delivery_df['ball']
delivery_df['balls_left'] = 120 - delivery_df['balls_bowled']
delivery_df['balls_left'] = delivery_df['balls_left'].clip(lower=0) 
delivery_df['overs_left'] = delivery_df['balls_left'] / 6
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype(int)
```

```python
delivery_df['wickets_fallen'] = delivery_df.groupby('match_id')['player_dismissed'].cumsum()
delivery_df['wickets_left'] = 10 - delivery_df['wickets_fallen']
```

```python
delivery_df['result'] = (delivery_df['batting_team'] == delivery_df['winner']).astype(int)
```

```python
delivery_df['crr'] = (delivery_df['current_score'] * 6) / delivery_df['balls_bowled']
delivery_df['rrr'] = np.where(delivery_df['balls_left'] > 0, 
                              (delivery_df['runs_left'] * 6) / delivery_df['balls_left'], 
                              0)
```

```python
final_df = delivery_df[['match_id','batting_team', 'bowling_team', 
                        'over', 'ball', 'current_score', 'target_score', 'runs_left', 
                        'balls_left', 'overs_left', 'crr', 'rrr', 'wickets_fallen', 'wickets_left', 
                        'result']].copy()
delivery_df['balls_left'] = 120 - delivery_df['balls_bowled']
delivery_df['balls_left'] = delivery_df['balls_left'].clip(lower=0)
```

```python
final_df
```

```python
features_to_keep = ['batting_team', 'bowling_team', 'target_score', 'runs_left', 
                    'balls_left', 'crr', 'rrr', 'wickets_left', 'result']
```

```python
df_model = final_df[features_to_keep].copy()
```

```python
df_encoded = pd.get_dummies(df_model, columns=['batting_team', 'bowling_team'], dtype=int)
```

```python
df_encoded
```

```python
X = df_encoded.drop(columns=['result'])
y = df_encoded['result']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
def rename_team_cols(col_name):
    if col_name.startswith('batting_team_'):
        
        return col_name.replace('batting_team_', 'bat_')
    elif col_name.startswith('bowling_team_'):
       
        return col_name.replace('bowling_team_', 'bowl_')
    return col_name

X_train.columns = [rename_team_cols(col) for col in X_train.columns]
X_test.columns = [rename_team_cols(col) for col in X_test.columns]


print("Cleaned Column Names:")
print(list(X_train.columns))


X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
```

```python
X_train=pd.read_csv("X_train.csv")
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
import time
```

```python
xgb_model = xgb.XGBClassifier(eval_metric='logloss', random_state=42)
```

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5],                
    'learning_rate': [0.05, 0.1],
    'reg_alpha': [0, 0.5, 1],           
    'reg_lambda': [1, 5, 10],            
    'gamma': [0, 0.1, 0.5]               
}
```

```python
grid_search = GridSearchCV(
    estimator=xgb_model, 
    param_grid=param_grid, 
    cv=3,                 
    scoring='accuracy',   
    n_jobs=-1,            
    verbose=2             
)
```

```python
print("Starting Hyperparameter Tuning. This may take a few minutes...")
start_time = time.time()

grid_search.fit(X_train, y_train)

end_time = time.time()
print(f"\nTuning completed in {end_time - start_time:.2f} seconds.")
print(f"Best Parameters Found: {grid_search.best_params_}")
print(f"Best Cross-Validation Accuracy: {grid_search.best_score_:.4f}")
```

```python
best_xgb_model = grid_search.best_estimator_
best_xgb_model.save_model('xgb_win_prob_model.json')
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
y_pred = best_xgb_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"\n=== Model Evaluation ===")
print(f"Overall Accuracy: {accuracy * 100:.2f}%\n")
```

```python
print(classification_report(y_test, y_pred, target_names=['Loss (0)', 'Win (1)']))
```

```python
print("=== Confusion Matrix ===")
conf_matrix = confusion_matrix(y_test, y_pred)
print(pd.DataFrame(conf_matrix, 
                   index=['Actual Loss', 'Actual Win'], 
                   columns=['Predicted Loss', 'Predicted Win']))
```
